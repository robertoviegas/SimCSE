```python
import ast
import sys
import pandas as pd
final = pd.read_excel('teste.xlsx')
for tema in tqdm(temas):
    if tema not in final['tema'].unique():
        try:
            response = client.models.generate_content(
                    model="gemini-2.0-flash", contents=f"você criará um dicionario python adaptavel para pandas dataframe de 50 linhas com o tema: {tema}.  Ele terá as seguintes colunas: st1, st2 e contradicion. A regra principal é que nenhuma frase pode ser igual a outra e precisam ter pelo menos uma palavra diferente entre elas. A frase A em st1 é pode ser gerada por diferentes niveis de complexidade na lingua portuguese nos mais variados temas. Dada a frase A. Crie uma frase B em st2 que possua uma relação de entailment com A: A sentença A implica a sentença B se, sempre que A for verdadeira, B também deve ser verdadeira. A frase B que voce gerará precisa ter complexidade e quantidade de caracteres semelhantes a A. Gere uma terceira C na coluna contradiction que possua semelhança vocabularia com as duas frases A e B, mas significado contraditório em relação às duas. Não utilize frases simples, forneça frases complexas para cada coluna. Forneca apenas o dicionario em string"
                )
            
            string_de_codigo = response.text.strip()

            # Passo 1: Isolar a string que representa o dicionário.
            # Encontramos o índice do primeiro '{' e do último '}' para extrair apenas o dicionário.
            inicio = string_de_codigo.find('{')
            fim = string_de_codigo.rfind('}') + 1
            string_do_dicionario = string_de_codigo[inicio:fim]

            # Passo 2: Converter a string do dicionário em um objeto de dicionário Python.
            # Usamos ast.literal_eval para fazer isso de forma segura.
            dados_extraidos = ast.literal_eval(string_do_dicionario)

            # # Passo 3: Criar o DataFrame do Pandas a partir do dicionário extraído.
            df = pd.DataFrame(dados_extraidos)
            df['tema'] = tema

            final =  pd.concat([final, df]).reset_index(drop=True)
            final.to_excel('teste.xlsx',index=False)
        except Exception as e:
            print(e)
            final.to_excel('teste.xlsx',index=False)
            sys.exit()
```
